[문제1]

파이썬 2.7 버전과 파이썬 3에서 range 차이를 설명해주세요

파이썬 2.7에서 range함수는 실제 리스트를 생성합니다. 즉 지정된 모든 값의 메모리를 저장하기 때문에 메모리 사용과 성능에 영향을 줍니다. 반면에 파이썬 3에서의 range는 리스트를 직접 반환하지 않고 반복가능한 객채만을 반환하기 때문에 메모리 효율성이 더 좋습니다.

[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요

얇은 복사는 새로운 복사본을 생성하지만 원본 객체의 요소를 반영하기 때문에 원본객체의 변경이 복사본에도 영향을 줄 수 있다.
깊은 복사는 완전히 독립적인 새로운 객체를 생성하기 때문에 원본 객체의 변경이 복사본에 영향을 주지 않는다.
