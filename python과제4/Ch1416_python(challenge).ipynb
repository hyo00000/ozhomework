다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"  
234567 → "23four5six7"  
10203 → "1zerotwozero3"    

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,   
혹은 바뀌지 않고 그대로인 문자열 s가 주어졌을때
s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

---
참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어  
0	zero  
1	one  
2	two  
3	three  
4	four  
5	five  
6	six  
7	seven  
8	eight  
9	nine  

---

```python
s = "77three4one"

def solution(s):
    words = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]    
    for i, word in enumerate(words):        
        s = s.replace(word, str(i))    
    return int(s)
            

result = solution(s)
print(result)

```

[문제5]

양의 정수 x가 하샤드 수이려면 x의 자릿수의 합으로 x가 나누어져야 합니다.   
예를 들어 18의 자릿수 합은 1+8=9이고, 18은 9로 나누어 떨어지므로 18은 하샤드 수입니다.   
주어진 x를 이용해 x가 하샤드 수인지 아닌지 검사하는 코드를 작성해주세요

x = 17

----

입출력 예시
x	결과값  
10	true  
12	true  
11	false  
13	false  

```python
def harshard_number(x):
    digit_sum = sum(int(digit) for digit in str(x))

    if x % digit_sum == 0 :
        return True
    else:
        return False

print(harshard_number(17))
```

[문제7]

0부터 9까지의 숫자 중 일부가 들어있는 정수 배열 numbers가 주어집니다.   
numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 출력해주세요

numbers = [5,8,1,0,6,9]

```python
numbers = [5, 8, 1, 0, 6, 9]
all_numbers = set(range(10))

def missing_numbers(numbers):

    for num in numbers:
        if num in all_numbers:
            all_numbers.remove(num)
    
        missing_number_sum = sum(all_numbers)
    return missing_number_sum

print(missing_numbers(numbers))
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
붕어빵재고 = {"팥붕어빵":10, "슈크림붕어빵":8, "초코붕어빵":5}
주문={}

while True:
    종류 = input("붕어빵 맛을 골라주세요(팥, 슈크림, 초코, 주문취소:'q')")
    if 종류 == 'q':
        break
    if 종류 not in 붕어빵재고:
        print("해당 맛의 붕어빵은 없습니다. 다시 입력해주세요.")
        continue
    
    갯수 = input("붕어빵 갯수를 골라주세요")
    if 갯수 > 붕어빵재고[종류]:
        print("주문하신 붕어빵의 재고가 부족합니다.")
    else:
        주문[종류]=주문.get(종류, 0) + 갯수
        붕어빵재고[종류] -= 갯수

print("주문내용:", 주문)
print("업데이트된 내용:", 붕어빵재고)
```
