DBMS 특징이 아닌것을 골라주세요

- 데이터 무경성
- 데이터 보안성
- 데이터 일관성
- 데이터 의존성

정답 : 데이터 의존성

DBMS 특징  
1. 데이터 무결성 : 데이터의 정확성과 일관성을 보장한다.  
2. 데이터 독립성 : 데이터와 응용 프로그램 간의 독립성을 제공한다.(스키마 변경이 응용 프로그램에 영향을 미치지 않음)  
3. 데이터 보안 : 데이터베이스 접근 제어 및 인증을 통해 데이터 보안을 제공한다.  
4. 데이터 일관성 : 트랜잭션 관리 기능을 통해 데이터의 일관성을 유지한다. -> 따라서 동시성 제어와 무결성 유지가 
               가능해 다중 지원자에게 지원됨  
5. 백업 및 복구 : 주기적인 백업과 복구 절차를 통해 데이터 손실을 최소화한다.

[문제3]

데이터베이스의 구조를 정의하는 것을 골라주세요

- 인덱스
- 트랜잭션
- 스키마
- SQL

정답 : 스키마

1.스키마 : 데이터베이스의 구조를 정의하는 것, 테이블, 행, 열, 인덱스, 관계드의 데이터베이스 구조를 명시함  
2.트랜잭션 : 테이터 베이스의 상태를 변화시키는 하나의 작업 단위, 데이터 무결성을 유지하는데 중요한 역할을 함  
3.인덱스 : 데이터베이스에서 데이터 검색 속도를 빠르게 하기 위해 사용되는 객체, 인덱스는 특정 열에 대한 포인터를   
          포함하여 데이터 검색시간을 단축시킴  
4.SQL : 데이터베이스에서 데이터를 관리하기 위해 사용되는 표준 프로그래밍 언어임, SQL을 사용하여 데이터베이스에  
        질의, 업데이트, 삭제 등의 작업을 수행할 수 있음.  
5.데이터베이스 관리 시스템(DBMS) : 데이터베이스를 생성하고 관리하는 소프트웨어  
                             MySQL, PostgreSQL, Oracle, Microsoft SQL Server등이 있음.

- 실시간 데이터 처리의 어려움
- 운영 비용이 비쌈

정답 : 실시간 데이터 처리의 어려움

<단점>
1. 대용량 테이터 처리나 복잡한 퀘리에 대한 성능이 다른 모델에 비해 낮을 수 있음 -> 대용량 성능이 낮음  
2. 수직적 확장은 한계가 있고 수평적 확장은 복잡하고 비용이 많이 듬  -> 확장 어려움  
3. 스키마 변경이 어려워서 요구사항 변경에 대응하기 어려움  -> 유연성 부족  
4. 많은 테이블 간의 조인은 성능에 부정적인 영향을 줌  -> 테이블 join의 복잡성  
5.서버 하드웨어 및 라이선스 비용이 상대적으로 높음 -> 고비용  
:RDBMS는 데이터의 구조화와 안정성 측면에서 강력해서 일반적인 업무 응용에 적합하다. 
but 성능이나 유연성등의 측면에서 고려해야 할 부분도 있다.

[문제8]

SQL 언어에 해당하지 않는 것을 골라주세요

- 데이터 처리 언어(DML)
- 데이터 제어 언어(DCL)
- 데이터 연결 언어(DCL)
- 데이터 정의 언어(DDL)

정답 : 데이터 연결 언어(DCL)

1. 데이터 정의 언어(DDL-Data Definition Language)
 : DDL은 데이터 베이스의 구조를 정의하고 관리하는데 사용된다.
   주요 명령어(CREATE/ALTER/DROP/TRUNCATE)

2. 데이터 처리 언어(DML-Data Manipulation Language)
 : DML은 데이터 검색, 삽입, 수정, 삭제하는 데 사용된다.
   주요 명령어(SELECT/INSERT/UPDATE/DELETE)

3. 데이터 제어언어(DCL-Data Control Language)
 : DCL은 데이터베이스에 대한 액세스를 제어하는데 사용된다.
   주요 명령어(GRANT/REVOKE)
